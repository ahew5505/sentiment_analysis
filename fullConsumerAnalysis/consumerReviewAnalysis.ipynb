# Calculating Pearson Correlation
I would like to explore a few areas that I think may correlate:
- Income and amount spent on speakers
- Sound Quality Rating and Satisfaction
- Sound Quality Rating and Usage
- Age and Usage
- Income and Price importance

```python
# Since this is a truncated data set, it may not contain all the values included
# in the survey responses. Leaving this unaddressed will lead to keyValue errors
# in later processes.
print(df["Sound Quality Rating"].unique())
print(df["Usage Frequency"].unique())
print(df["Amount Spent:"].unique())
print(df["How old are you?"].unique())
print(df["What's your annual household income?"].unique())
```

```python
#Starting by quantifying some non-numerical survey results
#For this truncated dataset, we can only include those results that are present.
sound_quality_map = {"Excellent" : 5,
              "Good" : 4,
              "Average" : 3,
              #"Poor" : 2,
              #"Very poor" : 1
                }

usage_frequency_map = {"Daily" : 5,
                       "Several times a week" : 4,
                       "Once a week" : 3,
                       "1-3 times a month" : 2,
                       "Rarely (fewer than once a month)" : 1}

income_map = {"More than $100,000" : 5,
                        "$75,000-$100,000" : 4,
                        "$50,000-$75,000" : 3,
                        "$25,000-$50,000" : 2,
                        "Less than $25,000" : 1,
                        "Prefer not to say" : 0}

age_map = {#"65 and over" : 7,
           #"55-64" : 6,
           #"45-54" : 5,
           "35-44" : 4,
           "25-34" : 3,
           "18-24" : 2,
           #"Under 18" : 1
        }

spending_map = {#"- More than $300" : 5,
                #"$200 to $300" : 4,
                "$100 to $200" : 3,
                "$50 to $100" : 2,
                "Less than $50" : 1}
```

```python
#We also definitely want to exclude results with "prefer not to say" as their income, as this doesn't yield meaninful insights
df_corr = df[df["What's your annual household income?"] != "Prefer not to say"]
```

```python
#Map values to columns
df_corr["Sound Quality Rating"] = df_corr["Sound Quality Rating"].map(sound_quality_map)
df_corr["Usage Frequency"] = df_corr["Usage Frequency"].map(usage_frequency_map)
df_corr["Amount Spent:"] = df_corr["Amount Spent:"].map(spending_map)
df_corr["How old are you?"] = df_corr["How old are you?"].map(age_map)
df_corr["What's your annual household income?"] = df_corr["What's your annual household income?"].map(income_map)
```

```python
#Time to find correlations!

income_to_spending = df_corr["What's your annual household income?"].corr(df_corr["Amount Spent:"])
income_to_price_importance = df_corr["What's your annual household income?"].corr(df_corr["Price Importance"])
sound_to_satisfaction = df_corr["Sound Quality Rating"].corr(df_corr["Satisfaction:"])
sound_to_usage = df_corr["Sound Quality Rating"].corr(df_corr["Usage Frequency"])
age_to_usage = df_corr["How old are you?"].corr(df_corr["Usage Frequency"])

print("Correlations between:\nIncome and Spending: {}\nIncome and Price Importance: {}\nSound Quality and Satisfaction: {}\nSound Quality and Usage Frequency: {}\nAge and Usage Frequency: {}".format(income_to_spending, income_to_price_importance, sound_to_satisfaction, sound_to_usage, age_to_usage))
```
